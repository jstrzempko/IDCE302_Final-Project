## Final Project - Script 2

The survey is provided in a text file and contains the names of individuals in a humanitarian organization and their picks for the county in Kenya in which they think operations should continue despite budget constraints. This script uses string manipulation techniques, custom functions, for loops, and if statements to coerce the data into a more understandable, standard format and extract relevant information described below in the outputs. 

**Inputs:**  

* KEcounty_votes.txt
* In Function 3: argument with name of a particular county to count votes for

* Function 1 (clean): a cleaned data set with voter name and county
* Function 2 (alreadyVoted): output statements if a person has already voted
* Function 3 (print_count): number of votes cast for a particular county
* Function 4 (count_votes): number of votes cast for each county
* Function 5 (bonus_max): output statement giving the county with the maximum number of votes as well as the vote count

```
#Import KEcounty_votes.txt from local computer
#Will bring up a browse button for you to browse to the location of the file on your computer
from google.colab import files
uploaded = files.upload()

#Function 1: Clean up/Munge the data

#Define function clean with parameter filename
def clean(filename):
  #Open and access the file
  with open(filename) as file:    
    #For every line in the file
    for line in file:
      #Remove extraneous spaces
      line = line.strip()
      #Use multiple assignment to assign variables to corresponding list items
      #Name (at index position 0) is name of the voter
      #Vote (at index position 1) is the county they voted for
      name, vote = line.split(" - ")
      #Clean the voter's name - strip of spaces, capitalize, replace double spaces with single
      name = name.strip().capitalize().replace("  "," ")
      #Clean the county votes
      #Remove extraneous spaces, capitalize all letters, replace double with single spaces
      #And fix the spelling on the Mombasa vote to be counted with the Mombassa votes
      vote = vote.strip().upper().replace("  ", " ").replace("MOMBASSA", "MOMBASA")
      #print statement to demonstrate output
      #print(name, 'voted for', vote)

#Call the function
clean("KEcounty_votes.txt")

#Function 2: Check if someone has voted before

#Define function alreadyVoted with parameter filename
def alreadyVoted(filename):
  #Create an empty list named voted
  #Will contain the names of everyone who voted
  voted = []
  #Open and access the file
  with open(filename) as file:    
    #For every line in the file
    for line in file:
      #Remove extraneous spaces
      line = line.strip()
      #Use multiple assignment to assign variables to corresponding list items
      #Name (at index position 0) is name of the voter
      #Vote (at index position 1) is the county they voted for
      name, vote = line.split(" - ")
      #Clean the voter's name - strip of spaces, capitalize, replace double spaces with single
      name = name.strip().capitalize().replace("  "," ")
      #Check if person has already voted, referencing empty list voted above
      if name in voted:
        #If already voted, print their name with a relevant statement 
        print(name + " has already voted! You cannot vote more than once.")
      else:
        #If they haven't already voted, add their name to the voted list
        voted.append(name)

alreadyVoted("KEcounty_votes.txt")

#Function 3: returns the number of votes for a particular county

#Define function print_votes with parameter filename and county_name
def print_votes(filename, county_name):
  #Start counter at 0 
  count = 0
  #Print statement giving name of county to count votes for
  print("Counting votes for", county_name, ":")
  #Access and open file
  with open(filename) as file:
    #For every line in the file
    for line in file:
      #Remove extraneous spaces
      line = line.strip()
      #Use multiple assignment to assign variables to corresponding list items
      #Name (at index position 0) is name of the voter
      #Vote (at index position 1) is the county they voted for
      #Splits name and vote with " - " 
      name, vote = line.split(" - ")
      #Clean the county votes
      #Remove extraneous spaces, capitalize all letters, replace double with single spaces
      #And fix the spelling on the Mombasa vote to be counted with the Mombassa votes
      vote = vote.strip().upper().replace("  ", " ").replace("MOMBASSA", "MOMBASA")
      #Every time the county name appears, increment the count by 1
      if vote == county_name:
        count = count + 1
  #Print the county name with the number of votes
  print(county_name, 'has', count, 'votes')  

#Call the function using all caps on the county name argument
print_votes("KEcounty_votes.txt", "KISUMU")

#Function 4: print the number of votes cast for each county

#Define function count_votes with parameter filename
def count_votes(filename):
  #Create an empty dictionary named counts
  #Will associate county names with vote counts
  counts = {}
  #Open and access the file
  with open(filename) as file:    
    #For every line in the file
    for line in file:
      #Remove extraneous spaces
      line = line.strip()
      #Use multiple assignment to assign variables to corresponding list items
      #Name (at index position 0) is name of the voter
      #Vote (at index position 1) is the county they voted for
      name, vote = line.split(" - ")
      #Clean the county votes
      #Remove extraneous spaces, capitalize all letters, replace double with single spaces
      #And fix the spelling on the Mombasa vote to be counted with the Mombassa votes
      vote = vote.strip().upper().replace("  ", " ").replace("MOMBASSA", "MOMBASA")
      #Check if person has already voted, referencing empty list voted above
      if not vote in counts:
        #If vote isn't in dictionary counts, then add it and give it a value of 1
        counts[vote] = 1
      else:
        #If vote is in dictionary already, increment the vote count by 1
        counts[vote] += 1
    print('Voting Results:')
    #Gives county name and number of votes it received
    for name in counts:
        print(name + ": " + str(counts[name]))

#Calls on file to use for county names and vote counts
count_votes("KEcounty_votes.txt")

#Function 5: Bonus Challenge

#Define function bonus_max with parameter filename
def bonus_max(filename):
  #Create an empty dictionary named counts
  #Will associate county names with vote counts
  counts = {}
  #Import itemgetter for use in retrieving tuple values
  from operator import itemgetter 
  #Open and access the file
  with open(filename) as file:    
    #For every line in the file
    for line in file:
      #Remove extraneous spaces
      line = line.strip()
      #Use multiple assignment to assign variables to corresponding list items
      #Name (at index position 0) is name of the voter
      #Vote (at index position 1) is the county they voted for
      name, vote = line.split(" - ")
      #Clean the county votes
      #Remove extraneous spaces, capitalize all letters, replace double with single spaces
      #And fix the spelling on the Mombasa vote to be counted with the Mombassa votes
      vote = vote.strip().upper().replace("  ", " ").replace("MOMBASSA", "MOMBASA")
      #Check if person has already voted, referencing empty list voted above
      if not vote in counts:
        #If vote isn't in dictionary counts, then add it and give it a value of 1
        counts[vote] = 1
      else:
        #If vote is in dictionary already, increment the vote count by 1
        counts[vote] += 1
    #Create tuple from items function
    counts_tuple = counts.items()
    #Variable defined as the maximum of the counts_tuple
    #where the key of the iterable tuple is the 1st element (i.e. the value)
    #at index position 0 (the name of county)
    #Example code from stackoverflow URL in README
    highest_county = max(counts_tuple, key = itemgetter(1))[0]
    #Print results with maximum vote value associated with corresponding county name 
    print ("The County with the Highest Score of", max(counts.values()) , "is: " + highest_county)

bonus_max("KEcounty_votes.txt")
```
